```python
import pandas as pd
data = pd.read_csv('Data.csv')
data = data.rename(columns={'AirCond': 'AirConditioner', 'HsKpngServ': 'Housekeeping','LndryServ':'Laundry','WashingMac':'WashingMachine','DryerMac':'DryerMachine','SwmPool':'SwimmingPool'})

spl=data.groupby('Category')
dh=spl.get_group('House').reset_index()
da=spl.get_group('Apartment').reset_index()
dr=spl.get_group('Room').reset_index()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=dh.corr(method='spearman')['Renting_price'].drop(labels=['index', 'Renting_price'])
y=dh.drop(['Property_name','Category','index','Renting_price'],axis=1).columns.values
plt.scatter(x,y)
plt.title('Correlation Coefficient Between Factors and House Renting Price',size=16)
plt.xlabel('Corr Rate',size=14)
plt.ylabel('Factors',size=14)
plt.scatter(x.max(),'Bedroom',color='red')
plt.show()
print('Corr with Bedroom: ',x.max())
```

```python
x=da.corr(method='spearman')['Renting_price'].drop(labels=['index', 'Renting_price'])
y=da.drop(['Property_name','Category','index','Renting_price'],axis=1).columns.values
plt.scatter(x,y)
plt.title('Correlation Coefficient Between Factors and Apartment Renting Price',size=16)
plt.xlabel('Corr Rate',size=14)
plt.ylabel('Factors',size=14)
plt.scatter(x.max(),'Bedroom',color='red')
plt.show()
print('Corr with Bedroom: ',x.max())
```

```python
x=dr.corr(method='pearson')['Renting_price'].drop(labels=['index', 'Renting_price'])
y=dr.drop(['Property_name','Category','index','Renting_price'],axis=1).columns.values
plt.scatter(x,y)
plt.title('Correlation Coefficient Between Factors and Room Renting Price',size=16)
plt.xlabel('Corr Rate',size=14)
plt.ylabel('Factors',size=14)
plt.scatter(x.max(),'AirConditioner',color='red')
plt.show()
print('Corr with AirConditioner: ',x.max())
```
