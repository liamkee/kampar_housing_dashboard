```python
import requests as req
from bs4 import BeautifulSoup

url_page1 = 'https://flatfy.my/house-for-rent-kampar'
url_page2 = 'https://flatfy.my/house-for-rent-kampar?page=2'

page1 = req.get(url_page1)
page2 = req.get(url_page2)

print(page1)
print(page2)

soup1 = BeautifulSoup(page1.text, 'html.parser')
soup2 = BeautifulSoup(page2.text, 'html.parser')
```

### Create Dataframe

```python
import pandas as pd

Data = pd.DataFrame()
```

### Get properties name

```python
name = soup1.find_all("a",class_="realty-preview__title-link")
name2 = soup2.find_all("a",class_="realty-preview__title-link")
name_list=[]

for i in name:
    name_list.append(i.get_text())

for i in name2:
    name_list.append(i.get_text())

Data['Prop_name'] = name_list
```

### Get price

```python
price = soup1.find_all("div", class_="realty-preview__price")
price2 = soup2.find_all("div", class_="realty-preview__price")
price_list=[]
for i in price:
    price_list.append(i.get_text())

for i in price2:
    price_list.append(i.get_text())

Data['Prop_price']=price_list
```

```python
rooms = soup1.find_all("div",class_="realty-content-layout__rooms-row")
rooms2 = soup2.find_all("div",class_="realty-content-layout__rooms-row")
rooms_list=[]
for i in rooms:
    rooms_list.append(i.get_text())

for i in rooms2:
    rooms_list.append(i.get_text())

Data['Number_of_Rooms']=rooms_list
```

```python
sqft = soup1.find_all("div",class_="realty-content-layout__properties-row")
sqft2 = soup2.find_all("div",class_="realty-content-layout__properties-row")
sqft_list=[]

for i in sqft:
    sqft_list.append(i.get_text())

for i in sqft2:
    sqft_list.append(i.get_text())
    
Data['Size']=sqft_list
```

```python
address = soup1.find_all("a",class_="realty-preview__origin-site")
address2 = soup2.find_all("a",class_="realty-preview__origin-site")
address_list=[]

for i in address:
    address_list.append(i.get_text())

for i in address2:
    address_list.append(i.get_text())
    
Data['Location']=address_list
```
