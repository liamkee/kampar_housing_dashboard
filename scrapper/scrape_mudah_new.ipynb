```python
import requests as req
from bs4 import BeautifulSoup

url_page1 = 'https://www.mudah.my/Perak/Properties-2002/kampar-for-rent?lst=0&fs=1&w=106&cg=2002&q=kampar&sa=178&so=1&st=u'
url_page2 = 'https://www.mudah.my/Perak/Properties-2002/kampar-for-rent?o=2&sa=178&q=kampar&so=1&th=1'

page1 = req.get(url_page1)
page2 = req.get(url_page2)

print(page1)
print(page2)

soup1 = BeautifulSoup(page1.text, 'html.parser')
soup2 = BeautifulSoup(page2.text, 'html.parser')

import pandas as pd

Property_name=[]
Renting_price=[]
Area=[]
Category=[]
Size=[]
Bedroom=[]
Bathroom=[]
Detail=[]
total1 = len(soup1.find_all("div",class_="listing_params_container"))
cBlock1 = soup1.find("div",class_="TabContainer")
block1 = cBlock1.find_next("div",class_="listing_params_container")

total2 = len(soup2.find_all("div",class_="listing_params_container"))
cBlock2 = soup2.find("div",class_="TabContainer")
block2 = cBlock2.find_next("div",class_="listing_params_container")

def get_info(b):#get data
    Property_name.append(b.find("h2",class_="list_title").text)
    Renting_price.append(b.find("div",class_="ads_price").text)
    Area.append(b.find("div",class_="area").text)
    Category.append(b.find("div",title="Category").text)
    Size.append(b.find("div",title="Size").text)
    
    try:
        Bedroom.append(b.find("div",title="Bedrooms").text)
    except:
        Bedroom.append("None")
    try:
        Bathroom.append(b.find("div",title="Bathrooms").text)
    except:
        Bathroom.append("None")
    
    #find for "more details"
    for n in b.find_all('a'):
        url_inside = n.get('href')
        page_inside = req.get(url_inside)
        soup_inside = BeautifulSoup(page_inside.text, 'html.parser')
        try:
            Detail.append(soup_inside.find("p",class_="moreless").text)
        except:
            Detail.append("None")
        
info = {'Property_name':Property_name,
    'Renting_price':Renting_price,
    'Area':Area,
    'Category':Category,
    'Size':Size,
    'Bedroom':Bedroom,
    'Bathroom':Bathroom,
    'Property_Details':Detail}
    
for i in range(total1):
    if (i==0):
        get_info(block1)
    else:
        block1 = block1.find_next("div",class_="listing_params_container")
        get_info(block1)

for i in range(total2):
    if (i==0):
        get_info(block2)
    else:
        block2 = block2.find_next("div",class_="listing_params_container")
        get_info(block2)
        
Data = pd.DataFrame(info)
Data.to_csv('Data.csv',index=False)
Data
```
